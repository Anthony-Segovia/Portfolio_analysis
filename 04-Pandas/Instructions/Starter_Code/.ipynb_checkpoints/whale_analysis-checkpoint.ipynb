# Data Cleaning

In this section, you will need to read the CSV files into DataFrames and perform any necessary data cleaning steps. After cleaning, combine all DataFrames into a single DataFrame.

Files:

* `whale_returns.csv`: Contains returns of some famous "whale" investors' portfolios.

* `algo_returns.csv`: Contains returns from the in-house trading algorithms from Harold's company.

* `sp500_history.csv`: Contains historical closing prices of the S&P 500 Index.

## Whale Returns

Read the Whale Portfolio daily returns and clean the data

```python
# Reading whale return
whale_returns_csvpath = Path("../Starter_Code/Resources/whale_returns.csv")
whale_returns_df =pd.read_csv(whale_returns_csvpath,index_col="Date",parse_dates=True,infer_datetime_format=True)
whale_returns_df.head(10)
```

```python
# Count nulls
whale_returns_df.isnull().sum()
```

```python
# Drop nulls
whale_returns_df = whale_returns_df.dropna().copy()
whale_returns_df
```

## Algorithmic Daily Returns

Read the algorithmic daily returns and clean the data

```python
# Reading algorithmic returns
algo_returns_csvpath = Path("../Starter_Code/Resources/algo_returns.csv")
algo_returns_df = pd.read_csv(algo_returns_csvpath, index_col = "Date", parse_dates= True, infer_datetime_format=True)
algo_returns_df.head(10)
```

```python
# Count nulls
algo_returns_df.isnull().sum()
```

```python
# Drop nulls
algo_returns_df = algo_returns_df.dropna().copy()
algo_returns_df
```

## S&P 500 Returns

Read the S&P 500 historic closing prices and create a new daily returns DataFrame from the data. 

```python
# Reading S&P 500 Closing Prices
sp500_history_csvpath = Path("../Starter_Code/Resources/sp500_history.csv")
sp500_history_df = pd.read_csv(sp500_history_csvpath, index_col = "Date", parse_dates= True, infer_datetime_format=True)
sp500_history_df.head(10)
```

```python
# Check Data Types
sp500_history_df.dtypes
```

```python
# Fix Data Types
sp500_history_df["Close"] = sp500_history_df["Close"].str.replace("$", "")
```

```python
sp500_history_df["Close"] = sp500_history_df["Close"].astype("float")
sp500_history_df.dtypes
```

```python
# Calculate Daily Returns
sp500_daily_returns = sp500_history_df.pct_change().dropna()
sp500_daily_returns.head(10)
```

```python
# Drop nulls
sp500_history_data_frame = sp500_history_df.dropna().copy()
sp500_history_data_frame
```

```python
# Rename `Close` Column to be specific to this portfolio.
sp500_daily_returns.rename(columns={'Close':'SP500_Close'}, inplace = True)
sp500_daily_returns.head(10)
```

## Combine Whale, Algorithmic, and S&P 500 Returns

```python
# Join Whale Returns, Algorithmic Returns, and the S&P 500 Returns into a single DataFrame with columns for each portfolio's returns.
combined_returns = pd.concat([sp500_daily_returns,algo_returns_df,whale_returns_df], axis='columns',join='inner')
combined_returns.sort_index(inplace=True)
combined_returns.head()
```

# Conduct Quantitative Analysis

In this section, you will calculate and visualize performance and risk metrics for the portfolios.

## Performance Anlysis

#### Calculate and Plot the daily returns.

```python
# Plot daily returns of all portfolios
combined_daily_returns = combined_returns.pct_change()
combined_daily_returns.plot(figsize=(15,10))
```

#### Calculate and Plot cumulative returns.

```python
# Calculate cumulative returns of all portfolios
sp500_cumulative_returns = (1 + sp500_daily_returns).cumprod()
algo_cumulative_returns = (1 + algo_returns_df).cumprod()
whale_cumulative_returns = (1 + whale_returns_df).cumprod()

# Plot cumulative returns

cumulative_returns = pd.concat([sp500_cumulative_returns,algo_cumulative_returns,whale_cumulative_returns],axis='columns',join='inner')
cumulative_returns.sort_index(inplace=True)
cumulative_returns.plot(figsize=(15,10), title = ("Cumulative Returns of All Portfolios"))
```

Determine the _risk_ of each portfolio:

### Create a box plot for each portfolio

### Calculate and Plot Beta for a chosen portfolio and the S&P 500

```python
# Calculate covariance of a single portfolio
algo_covariance = combined_returns['Algo 1'].cov(combined_returns["SP500_Close"])

rolling_algo_covariance = combined_returns['Algo 1'].rolling(window=21).cov(combined_returns["SP500_Close"])
rolling_algo_covariance.plot(figsize=(15, 10), title='Rolling 21-Day Covariance of Algo 1 Returns vs. S&P 500 Returns')
```

```python
# Calculate variance of S&P 500
rolling_variance = combined_returns["SP500_Close"].rolling(window=21).var()
rolling_variance.plot(figsize=(15, 10), title='Rolling 21-Day Variance of S&P 500 Returns')
```

```python
# Calculate 21-day rolling beta of Algo 1
rolling_beta = rolling_algo_covariance / rolling_variance
rolling_beta.plot(figsize=(15, 10), title='Rolling 21-Day Beta of Algo 1')
```

# Sharpe Ratios
In reality, investment managers and thier institutional investors look at the ratio of return-to-risk, and not just returns alone. After all, if you could invest in one of two portfolios, and each offered the same 10% return, yet one offered lower risk, you'd take that one, right?

### Using the daily returns, calculate and visualize the Sharpe ratios using a bar plot

```python
# Annualized Sharpe Ratios
sharpe_ratios = (combined_returns.mean() * 252) / (combined_returns.std() * np.sqrt(252))
sharpe_ratios
```

```python
# Visualize the sharpe ratios as a bar plot
sharpe_ratios.plot(kind="bar", title="Sharpe Ratios")
```

### Calculate and Plot Rolling 60-day Beta for Your Portfolio compared to the S&P 500

```python
# Create a new pivot table where the columns are the closing prices for each ticker
combined_df = pd.concat([combined_custom_daily_returns, sp500_daily_returns], axis="columns", join="inner")

# Sort datetime index in ascending order (past to present)
combined_df.sort_index(inplace=True)

# Set column names to 'AMZN' and 'S&P 500'
combined_df.columns = ["DKNG","CHWY", "MARA", "S&P 500"]
combined_df.head()
```

```python
 # Calculate variance of all daily returns of DKNG vs. S&P 500
dkng_variance = combined_df["S&P 500"].var()

# Calculate covariance of all daily returns of DKNG vs. S&P 500
dkng_covariance = combined_df["DKNG"].cov(combined_df["S&P 500"])

# Calculate the beta
dkng_beta = combined_covariance / combined_variance
dkng_beta

# Calculate 30-day rolling covariance of AMZN vs. S&P 500 and plot the data
dkng_rolling_covariance = combined_df["DKNG"].rolling(window=60).cov(combined_df["S&P 500"])
dkng_rolling_covariance.plot(figsize=(20, 10), title="Rolling 30-Day Covariance of DKNG Returns vs. S&P 500 Returns")
```

```python
 # Calculate variance of all daily returns of DKNG vs. S&P 500
chwy_variance = combined_df["S&P 500"].var()

# Calculate covariance of all daily returns of DKNG vs. S&P 500
chwy_covariance = combined_df["CHWY"].cov(combined_df["S&P 500"])

# Calculate the beta
chwy_beta = combined_covariance / combined_variance
chwy_beta

# Calculate 30-day rolling covariance of AMZN vs. S&P 500 and plot the data
chwy_rolling_covariance = combined_df['DKNG'].rolling(window=60).cov(combined_df['S&P 500'])
chwy_rolling_covariance.plot(figsize=(20, 10), title='Rolling 30-Day Covariance of CHWY Returns vs. S&P 500 Returns')
```

```python
 # Calculate variance of all daily returns of DKNG vs. S&P 500
mara_variance = combined_df["S&P 500"].var()

# Calculate covariance of all daily returns of DKNG vs. S&P 500
mara_covariance = combined_df["MARA"].cov(combined_df["S&P 500"])

# Calculate the beta
mara_beta = combined_covariance / combined_variance
mara_beta

# Calculate 60-day rolling covariance of AMZN vs. S&P 500 and plot the data
mara_rolling_covariance = combined_df['MARA'].rolling(window=60).cov(combined_df['S&P 500'])
mara_rolling_covariance.plot(figsize=(20, 10), title='Rolling 30-Day Covariance of MARA Returns vs. S&P 500 Returns')
```
